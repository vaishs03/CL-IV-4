```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Load needed sheet (you can change the sheet name if needed)
sales_df = pd.read_excel("sales_data.xlsx")

# Preview 
display(sales_df.head())
```

```python
# --- Sales Data ---
sales_df.dropna(inplace=True)
sales_df.drop_duplicates(inplace=True)
sales_df['OrderDate'] = pd.to_datetime(sales_df['OrderDate'])
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=sales_df, x='OrderDate', y='Sale_amt', estimator='sum')
plt.title("Sales Over Time")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=sales_df, x='Region', y='Sale_amt')
plt.title("Sales by Region")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
